### Network Assignment

```python
import networkx as nx
import matplotlib.pyplot as plt

# Define nodes and edges for the network
pentagram_nodes = ['D', 'F', 'I', 'N', 'S']
green_nodes = ['BIH', 'GEO', 'ISR', 'MNE', 'SRB', 'CHE', 'TUR', 'UKR', 'GBR', 'AUS', 'HKG', 'USA']
yellow_nodes = ['AUT', 'BEL', 'BGR', 'HRV', 'CZE', 'EST', 'FRA', 'DEU', 'GRC', 'HUN', 'IRL', 'ITA', 'LVA', 'LUX', 
                'NLD', 'PRT', 'ROU', 'SVK', 'SVN', 'ESP']

G = nx.Graph()

# Create a pentagram (connecting the pentagram nodes to each other)
for i, node in enumerate(pentagram_nodes):
    G.add_edge(node, pentagram_nodes[(i + 1) % len(pentagram_nodes)])

# Connect each green and yellow node randomly to a pentagram node
import random
for node in green_nodes + yellow_nodes:
    G.add_edge(node, random.choice(pentagram_nodes))

# Define node colors
node_colors = []
for node in G.nodes():
    if node in pentagram_nodes:
        node_colors.append('blue')
    elif node in green_nodes:
        node_colors.append('green')
    elif node in yellow_nodes:
        node_colors.append('yellow')

# Set positions: Pentagram in the center and others outside
pentagram_pos = nx.shell_layout(G, nlist=[pentagram_nodes, green_nodes + yellow_nodes])

# Draw the network graph
plt.figure(figsize=(10, 8))
nx.draw(G, pos=pentagram_pos, with_labels=True, node_color=node_colors, node_size=1200, font_size=8, font_color='black')
plt.title('Network Graph with Pentagram in the Center', fontsize=14, fontweight='bold')
plt.show()
```
